#PERFORMING EDA
EDA is mandatory and important foundation for every project to understand the data more better way to present the project with pattern analysation..
so in this project the possible ways are
1. Customer Age Distribution
2. Income Distribution
3. Total Spending Distribution
4. Top Product Categories

lets interpret these respectively in successive blocks and followed by insights



```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=30, kde=True)

plt.title('Customer Age Distribution')
plt.show()
```

most of the customers age ranged between 40-50

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Income'], bins=30, kde=True)

plt.title('Income Distribution')
plt.show()
```

Customer income varies significantly across segments.

```python
plt.figure(figsize=(8,5))
sns.histplot(df['MntTotal'], bins=30, kde=True)

plt.title('Total Customer Spending')
plt.show()
```

A small group of customers contributes a large portion of revenue.

```python
products = [
    'MntWines',
    'MntFruits',
    'MntMeatProducts',
    'MntFishProducts',
    'MntSweetProducts',
    'MntGoldProds'
]

df[products].sum().sort_values().plot(
    kind='barh',
    figsize=(10,6)
)

plt.title('Total Sales by Product Category')
plt.show()
```

Wine and meat products generate the highest revenue.

```python
channels = [
    'NumWebPurchases',
    'NumCatalogPurchases',
    'NumStorePurchases'
]

df[channels].sum().plot(
    kind='bar',
    figsize=(8,5)
)

plt.title('Purchases by Channel')
plt.show()
```

Store purchases dominate customer buying behavior

#CUSTOMER SEGMENTATION
Customer Segmentation means dividing customers into different groups based on similar characteristics or behavior.
In our project we have 5 factors to analyse
1. Income Segmentation
2. Spending by Income Group
3. Family Size Analysis
4. Marketing Campaign Analysis
5. Correlation Heatmap

lets interpret these respectively in successive blocks and followed by insights

```python
df['Income_Group'] = pd.qcut(
    df['Income'],
    q=4,
    labels=[
        'Low Income',
        'Middle Income',
        'Upper Middle',
        'High Income'
    ]
)
```

qcut() is a Pandas function used to divide numerical data into equal-sized groups

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=df,
    x='Income_Group',
    y='MntTotal'
)

plt.title('Average Spending by Income Group')
plt.xticks(rotation=15)
plt.show()
```

Higher-income customers spend significantly more

```python
df['Family_Size'] = (
    df['Kidhome'] +
    df['Teenhome'] + 1
)

sns.boxplot(
    x='Family_Size',
    y='MntTotal',
    data=df
)

plt.title('Family Size vs Spending')
plt.show()
```

Smaller families generally spend more per customer

```python
campaigns = [
    'AcceptedCmp1',
    'AcceptedCmp2',
    'AcceptedCmp3',
    'AcceptedCmp4',
    'AcceptedCmp5'
]

df[campaigns].sum().plot(
    kind='bar',
    figsize=(8,5)
)

plt.title('Campaign Acceptance')
plt.show()
```

Campaign acceptance rates vary considerably.

```python
plt.figure(figsize=(14,10))

sns.heatmap(
    df.corr(numeric_only=True),
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')
plt.show()
```
